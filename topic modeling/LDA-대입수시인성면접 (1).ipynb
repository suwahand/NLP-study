```python
tokenizer =Mecab()
text = "14년이라는 시간동안 운동을 했어 "
print(tokenizer.morphs(text))
```

```python
def preprocess_text(text):

    text = re.sub(r'\([^)]*\)', r'', text)#소괄호 내용 삭제
    text = re.sub(r'\[[^]]*\)', r'', text)#대괄호 내용 삭제
    text = re.sub(r"[^ㄱ-ㅎㅏ-ㅣ가-힣]+", "", text)#한글 이외에 숫자, 특수문자, 영어 제거
    text = re.sub(r'[" "]+', " ", text) #연속된 공백 하나의 공백으로 처리
    return text
```

```python
df['question1'] = df['question1'].apply(lambda x : preprocess_text(str(x)))
df['question1'][:10]
```

```python
#불용어 사전
stopwords = pd.read_table('stopword.txt', sep=',')
stopwords
```

```python
#토큰화 및 불용어 처리
tokenized_data = []
for x in df['question1']:
    temp = tokenizer.nouns(x) #명사만 추출
    temp = [x for x in temp if len(x)>1]
    temp = [x for x in temp if x not in stopwords]
    tokenized_data.append(temp)
    
    
tokenized_data
```

```python
#LDA모델 훈련
from gensim import corpora
dictionary= corpora.Dictionary(tokenized_data)
corpus = [dictionary.doc2bow(text) for text in tokenized_data]
print(corpus[:5])
```

```python
len(dictionary)
```

```python
#LDA모델 훈련
import gensim
NUM_TOPICS = 10
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15 )
topics = ldamodel.print_topics(num_words=6)
for num, topic in topics:
    topics = sorted(topics, key = lambda x:x[0])
    num = num+1
    print(f"Topic#{num}:{topic}")
    topics = topics
```
