```python
import pandas as pd
from konlpy.tag import Mecab
import re
from pandas import DataFrame
import openpyxl
```

```python
filename="대입수시인성면접.xlsx"
book = openpyxl.load_workbook(filename)
```

```python
df = pd.read_excel(filename)
df = df.fillna("")
print(df.isnull().sum())#
```

```python
qs2=df.iloc[:,1].tolist()  # 인덱스 아래의 있는 것들을 리스트화 [0]    
#질문번호 변경시 변경부분
qs2
```

```python
tokenizer =Mecab()
text = "14년이라는 시간동안 운동을 했어 "
print(tokenizer.morphs(text))
```

```python
def preprocess_text(text):
    text = re.sub(r'\([^)]*\)', r'', text) #소괄호 내용삭제
    text = re.sub(r'\[[^]]*\)', r'', text)#대괄호 삭제
    text = re.sub(r"[^ㄱ-ㅎㅏ-ㅣ가-힣.?!,' ]+","",text) #한글 및 지정 특수 문자 외 제거
    
    text = re.sub(r'["."]+'," ",text)#.제거
    text = re.sub(r'[" "]+'," ",text)#연속된 공백 하나로 처리
    return text
```

```python
preprocess_text("안녕.        sfl 123 (df)[sadf].,!?")
```

```python
preprocess_text(qs2[0])#질문번호 변경시 변경부분
```

```python
df_data = pd.DataFrame(qs2, columns = ['Answers'])#질문번호 변경시 변경부분
print(df_data)
```

```python
df_data['Answers'] = df_data['Answers'].apply(lambda x: preprocess_text(x))
df_data['Answers'][:10]
```

```python
script= df_data['Answers']
```

```python
#불용어 사전
f = open("stopword.txt", "r")

stopwords = []
while True:
    line = f.readline().strip()
    if not line: break
    stopwords.append(line)
print(stopwords)
```

```python
#토큰화와 불용어처리
tokenized_data = []
for x in df_data['Answers']:
    temp = tokenizer.nouns(x)
    temp = [x for x in temp if x not in stopwords]
    temp = [x for x in temp if len(x)>1]
    tokenized_data.append(temp)
```

```python
tokenized_data[0]
```

```python
#정수인코딩 및 빈도수 기록(인코딩 값/단어 빈도수)
from gensim import corpora
dictionary = corpora.Dictionary(tokenized_data)
corpus = [dictionary.doc2bow(text) for text in tokenized_data] #말뭉치 생성 벡터화
print(corpus[10])
```

```python
len(dictionary)
```

```python
#LDA모델 훈련
import gensim
NUM_TOPICS = 8
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word = dictionary, passes=15)
topics = ldamodel.print_topics(num_words=10)
for num, topic in topics:
    topics = sorted(topics, key = lambda x:x[0])
    num= num+1
    print(f"Topic# {num}:{topic}")
```

```python
df = pd.DataFrame(topics, columns = ['topic num', 'words'])
df['topic num']=df['topic num'].apply(lambda  x: x+1)
df['topic']=" "
df.head()
```

df.to_excel('대입수시_질문2(LDA).xlsx') #질문번호 변경시 변경부분 #한번 하고 나서는 markdown으로 변경

```python
def make_topictable_per_doc(ldamodel, corpus):
    global topic_table
    topic_table=pd.DataFrame()
    
    #데이터 프레임 부분
    for i, topic_list in enumerate(ldamodel[corpus]):
    
        doc = topic_list[0] if ldamodel.per_word_topics else topic_list
        doc = sorted(doc, key = lambda x: (x[1]), reverse=True)
        
        for j, (topic_num, prop_topic)in enumerate(doc):#몇 번토픽인지 비중 나눠서 저장
            if j ==0:
                topic_table = topic_table.append(pd.Series([int(topic_num+1), script[i], round(prop_topic, 4), topic_list]), ignore_index=True)
                #데이터 프레임형태 만듦(답변 내용과 토픽 부분들이 같이 나올 수 있도록)
            else:
                break
    return(topic_table)
```

```python
topictable = make_topictable_per_doc(ldamodel, corpus)
topictable = topictable.reset_index() #답변자번호 의미
topictable.columns = ['답변자', '가장 비중이 높은 토픽', '답변내용', '가장 높은 토픽의 비중', '각 토픽 비중']
topictable['LDA (1)']=" "
topictable['pyLDAvis (2)']=" "
topictable['적합한 버전']=" "
#토픽과 적합한 토픽은 채워나가야 하는 부분

topictable = topictable[['답변자', '가장 비중이 높은 토픽','LDA (1)','pyLDAvis (2)','답변내용','적합한 버전','가장 높은 토픽의 비중','각 토픽 비중']]
topictable
```
