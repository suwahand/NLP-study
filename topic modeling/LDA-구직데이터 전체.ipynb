```python
def preprocess_text(text):
    text = re.sub(r'\([^)]*\)', r'', text) #소괄호 내용삭제
    text = re.sub(r'\[[^]]*\)', r'', text)#대괄호 삭제
    text = re.sub(r"[^ㄱ-ㅎㅏ-ㅣ가-힣.?!,' ]+","",text) #한글 및 지정 특수 문자 외 제거
    text = re.sub(r'["oO"][3,]',r'', text) #익명화 부분 삭제 
    text = re.sub(r'["0-9"]',"", text) #토픽모델링이라서 숫자가 크게 상관 없을 거 같아 제거
    
    text = re.sub(r'["."]+'," ",text)#.제거
    text = re.sub(r'[" "]+'," ",text)#연속된 공백 하나로 처리
    return text
```

```python
df_data = pd.DataFrame(data1, columns = ['Answers'])
print(df_data)
```

```python
df_data['Answers'] = df_data['Answers'].apply(lambda x: preprocess_text(str(x)))
df_data['Answers'][:10]
```

```python
#불용어 사전
f = open("stopword.txt", "r")

stopwords = []
while True:
    line = f.readline().strip()
    if not line: break
    stopwords.append(line)
print(stopwords)
```

```python
#토큰화 및 불용어
tokenized_data = []
for x in df_data['Answers']:
    temp = tokenizer.nouns(x)
    temp = [ x for x in temp if x not in stopwords]
    tokenized_data.append(temp)
```

```python
tokenized_data
```

```python
#정수 인코딩 및 빈도수 기록
#인코딩 된 값/ 단어 빈도수
from gensim import corpora
dictionary = corpora.Dictionary(tokenized_data)
corpus = [dictionary.doc2bow(text) for text in tokenized_data]
print(corpus[5])
```

```python
len(dictionary)
```

```python
#LDA모델 훈련시키기

import gensim
NUM_TOPICS = 41
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=NUM_TOPICS, id2word = dictionary, passes=15)
topics = ldamodel.print_topics(num_words=6)
for topic in topics:
    print(topic)
```
