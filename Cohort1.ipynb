```python
#출처 :https://workingwithpython.com/pythoncohortanalysis/
import pandas as pd
from pandas import Series
from pandas import DataFrame

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('relay-foods.xlsx', sheet_name = 'Purchase Data - Full Study')
df.head()
```

```python
df.info()
#데이터의 구성과 각 모든 컬럼에 결측치가 없는 것을 확인
```

```python
df.drop(['CommonId', 'PupId', 'PickupDate', 'TotalCharges'],
           axis=1, inplace=True)
df.head()
```

```python
#주문일자(orderdate)단위 기간 만들기
df['OrderDate'].head()
```

```python
df['OrderDate'].dtype
#시계열데이터임을 알수 있음 
#시리즈의 dt accessor를 적용해 시계열 데이터 처리
```

```python
df['OrderPeriod']=df['OrderDate'].dt.strftime('%Y-%m')
df.head()
```

```python
#유저의 첫번째 주문을 토대로 유저별 코호트 그룹 설정
df.set_index('UserId', inplace=True)
df.head()
```

```python
FirstOrder=df.groupby(df.index)['OrderDate'].min()
FirstOrder.head()
```

```python
FirstOrder=FirstOrder.dt.strftime('%Y-%m')
FirstOrder.head()
#코호트 분석의 기준은 월 단위 이므로 월로 변경
```

```python
df['FirstOrder']=FirstOrder
df.reset_index(inplace=True)
df.head()
```

```python
#코호트 그룹과 주문기간으로 데이터 집계
#코호트 분석은 각각의 방문을 모두 집계하는 것이 아닌 순방문자로 집계함(1)
grouped = df.groupby(['FirstOrder', 'OrderPeriod'])
```

```python
cohorts=grouped['UserId'].nunique()
cohorts.head()
#집계함수로 unique값의 개수를 세는 nunique함수 사용
#고객이 여러번 방문해도 한 번으로 집계되어 월별 순 방문자수를 파악
```

```python
cohorts=cohorts.reset_index()
cohorts.head()
```

```python
cohorts.rename({'UserId':'TotalUsers'}, axis=1, inplace=True)
cohorts.head()
```

```python
each_period=cohorts['FirstOrder'].value_counts().sort_index()
each_period
#두번째만 14가 아닌 13인이유
```

```python
cohorts[cohorts['FirstOrder']=='2009-02']
#2010-03데이터가 orderperiod에 없음
```

```python
cohorts[cohorts['FirstOrder']=='2009-04'].tail()
```

```python
cohorts[cohorts['FirstOrder']=='2009-05'].tail()
#하지만 다른 데이터는 2010-03 존재
```

```python
cohorts.loc[cohorts.shape[0], 'FirstOrder']='2009-02'
cohorts.loc[cohorts.shape[0]-1, 'OrderPeriod']='2010-03'
cohorts.loc[cohorts.shape[0]-1, 'TotalUsers']=0
cohorts.sort_values(['FirstOrder', 'OrderPeriod'], inplace=True)
```

#### 데이터가 없는 기간도 포함하여 코호트 분석을 진행하기 위해 해당기간인 2009-02의 FirstOrder와 OrderPeriod를 입력후 totalusers에 0을 입력
#### 그리고 sort_values함수로 전체 데이터 순서 정렬
이후 다시 확인하면 올바로 2010-03이 추가된 것을 볼 수 있음
이에 대한 이유는 모종의 이유(서버 혹은 운영의 문제)라고 한다

```python
cohorts[cohorts['FirstOrder']=='2009-02']
```

```python
each_period=cohorts['FirstOrder'].value_counts().sort_index()
each_period
#잘 되는 모습
```

```python
#기간별 레이블 붙이기 
#코호트 분석에서는 처음 들어간 기간이 0으로 표시되며 이 데이터에도 숫자레이블을 붙여야 함
cohortperiod=[]

for x in each_period:
    for y in range(x):
        cohortperiod.append(y)
```

```python
cohorts['CohortPeriod']=cohortperiod
cohorts.head(20)
#cohortperiod가 잘 레이블링된 모습
```

```python
cohorts.set_index(['FirstOrder', 'CohortPeriod'], inplace=True)
cohorts.head()
```

```python
#이후 동일한 경험(firstorder)와 동일한 기간(cohortperiod)로 인덱스화
cohorts=cohorts['TotalUsers'].unstack(1)
cohorts
#unstack을 1로 하면 두번째 인덱스인 동일한 거리가 컬럼으로 변경되며
#y축은 코호트, x축은 기간으로 표현되는 코호트 분석의 형태를 가짐
```

```python
#코호트별 사용자 유지율 구하기
cohorts[0] #코호트 첫번째 기간(0주)에 대한 데이터
```

```python
user_retention=cohorts.divide(cohorts[0], axis=0)
user_retention
#전체 데이터(cohorts)를 첫 코호트 기간의 수방문자로 나누어주면 됨
#divide함수에 axis=0을 명시해야 전체 데이터를 첫 코호트 기간의 데이터로 나눌 수 있음
```

```python
#코호트 시각화
%matplotlib inline

import seaborn as sns
plt.figure(figsize=(12,8))
plt.title('Cohort Analysis', fontsize=19)
sns.heatmap(user_retention, annot=True, fmt='.0%') 
plt.show()
#annot은 각 데이터를 히트맵상에 표기할 것인지 불리언 결정
#fmt는 숫자 표시형식인데 .0%는 백분율로 하되 1의 자리까지 표현하라는 뜻
```
